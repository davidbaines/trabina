```python
def macula_to_usfm(sources, target):
    
    scripture_filename = Path ('Macula_XML') / source
    with scripture_filename.open("rb") as scripture_file:
        scripture_tree = etree.parse(scripture_file)

    usfm_filename = Path ('Macula') / f'{target}Macula.SFM'
    with usfm_filename.open("w", encoding="utf-8") as usfm_file:
        usfm_file.write(f'\id {target[2:]}\n')

        for chapter in scripture_tree.getroot().find("tanach").find("book").xpath("c"):
            usfm_file.write(f'\c {chapter.attrib["n"]}\n')

            for verse in chapter.xpath("v"):
                word_list = []
                for word in verse.xpath("w"):
                    word_list.append(word.text)
                text = ' '.join(word_list)
                usfm_file.write(f'\\v {verse.attrib["n"]} {text}\n')
    return
```
